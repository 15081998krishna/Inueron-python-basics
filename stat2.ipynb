```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import pylab

def plot_data(sample):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample,kde=True)
  plt.subplot(1,2,2)
  stats.probplot(sample, dist='norm',plot=pylab)
  plt.show()
```

```python
plot_data(data)
```

```python
plot_data(s)
```

```python


plot_data(np.log(s2))

```

```python


plot_data(s2)

```
